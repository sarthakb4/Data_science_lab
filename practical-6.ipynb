```python
# Experiment 06: Titanic Data Analysis and Visualization

# Step 1: Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load the Titanic dataset
df = pd.read_csv("Titanic-Dataset(1).xls")

# Step 3: Display first 5 records
print("First 5 Rows:")
display(df.head())

# Step 4: Display dataset info in tabular form
print("\nDataset Info (Tabular):")
info_df = pd.DataFrame({
    'Non-Null Count': df.count(),
    'Dtype': df.dtypes
})
display(info_df)

# Step 5: Check missing values
print("\nMissing Values Before Cleaning:")
print(df.isnull().sum())

# Step 6: Handle Missing Values (Corrected)

# Fill missing Age values
df['Age'] = df['Age'].fillna(df['Age'].median())

# Fill missing Embarked values
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

# Drop Cabin column
df.drop(columns=['Cabin'], inplace=True)

print("\nMissing Values After Cleaning:")
print(df.isnull().sum())

# Step 7: Data Visualization

# Create a new column for clearer labels in plots
df['Survived_Label'] = df['Survived'].map({0: 'Dead', 1: 'Survived'})

# 1 Survival Count
plt.figure()
sns.countplot(x='Survived_Label', data=df)
plt.title("Survival Count")
plt.xlabel("Survival Status") # Added for clarity
plt.show()

# 2 Survival by Gender
plt.figure()
sns.countplot(x='Sex', hue='Survived_Label', data=df)
plt.title("Survival by Gender")
plt.xlabel("Gender") # Added for clarity
plt.ylabel("Count") # Added for clarity
plt.show()

# 3 Survival by Passenger Class
plt.figure()
sns.countplot(x='Pclass', hue='Survived_Label', data=df)
plt.title("Survival by Passenger Class")
plt.xlabel("Passenger Class") # Added for clarity
plt.ylabel("Count") # Added for clarity
plt.show()

# 4 Age Distribution by Survival
plt.figure()
sns.histplot(data=df, x='Age', hue='Survived_Label', bins=30, kde=True, multiple='stack') # Added hue and multiple='stack'
plt.title("Age Distribution by Survival")
plt.xlabel("Age") # Added for clarity
plt.ylabel("Frequency") # Added for clarity
plt.legend(title='Survival Status') # Added legend title
plt.show()

print("\nTitanic Data Analysis Completed Successfully!")
```
