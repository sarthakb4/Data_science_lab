```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)

# Example usage:
number = int(input("Enter a number to compute its factorial: "))
result = factorial(number)
print(f"The factorial of {number} is {result}")
```

```python
# Get input from the user
num = int(input("Enter a number to display its multiplication table: "))

# Display the multiplication table
print(f"Multiplication Table for {num}:")
for i in range(1, 11): # Loop from 1 to 10
    print(f"{num} x {i} = {num * i}")
```

```python
# Get input from the user
number_str = input("Enter a number to check if it's a palindrome: ")

# Reverse the string
reversed_str = number_str[::-1]

# Check if it's a palindrome
if number_str == reversed_str:
    print(f"The number {number_str} is a palindrome.")
else:
    print(f"The number {number_str} is not a palindrome.")
```

```python
knowledge_base = [
    {
        "conditions": {
            "printer_on": False
        },
        "conclusion": "Check power and turn on printer."
    },
    {
        "conditions": {
            "printer_on": True,
            "connected_to_computer": False
        },
        "conclusion": "Check printer cable connection to computer."
    },
    {
        "conditions": {
            "printer_on": True,
            "connected_to_computer": True,
            "paper_in_tray": False
        },
        "conclusion": "Add paper to the printer tray."
    },
    {
        "conditions": {
            "printer_on": True,
            "connected_to_computer": True,
            "paper_in_tray": True,
            "ink_levels_low": True
        },
        "conclusion": "Replace ink cartridges."
    },
    {
        "conditions": {
            "printer_on": True,
            "connected_to_computer": True,
            "paper_in_tray": True,
            "ink_levels_low": False
        },
        "conclusion": "Printer should be working now."
    }
]

print("Knowledge base for printer diagnosis created.")
```

```python
def infer_diagnosis(knowledge_base):
    user_facts = {}

    # Collect all unique condition keys from the knowledge base
    all_conditions = set()
    for rule in knowledge_base:
        for condition_key in rule["conditions"]:
            all_conditions.add(condition_key)

    # Prompt user for each condition
    print("\nPlease provide information for the following:")
    for condition_key in all_conditions:
        while True:
            response = input(f"Is '{condition_key}' true? (yes/no): ").lower()
            if response == 'yes':
                user_facts[condition_key] = True
                break
            elif response == 'no':
                user_facts[condition_key] = False
                break
            else:
                print("Invalid input. Please enter 'yes' or 'no'.")

    print("\n--- Running Diagnosis ---")
    for rule in knowledge_base:
        conditions_met = True
        for condition_key, expected_value in rule["conditions"].items():
            if condition_key not in user_facts or user_facts[condition_key] != expected_value:
                conditions_met = False
                break

        if conditions_met:
            print(f"Diagnosis: {rule["conclusion"]}")
            return

    print("No specific diagnosis found. Please check all components carefully.")

# Example usage:
infer_diagnosis(knowledge_base)
```

```python
infer_diagnosis(knowledge_base)
```
